```python
import numpy as np
import pandas as pd
from dataclasses import dataclass
import requests
import time
import subprocess
import os
import csv
from IPython.display import clear_output
```

```python
df_train = pd.read_csv('data/train.csv')
df_train
```

```python
song_ids = df_train['song_id'].unique()
print("number of unique songs in train set: ", len(song_ids))
```

```python
df_songs = pd.read_csv('data/songs.csv')
df_songs
```

```python
relevant_songs = df_songs[df_songs['song_id'].isin(song_ids)]
relevant_songs # interestingly, there are a few songs from the training data that are not included in the songs data
```

```python
relevant_artist_names = relevant_songs['artist_name'].unique()
print("number of unique artists in relevant songs: ", len(relevant_artist_names))
```

```python
@dataclass
class ArtistInfo:
    artist_name: str
    gender: str
    country: str

def request_musicbrainz(endpoint, params, retries=5):
    url = f"https://musicbrainz.org/ws/2/{endpoint}"
    headers = {
        "User-Agent": "SchoolProject/1.0 (jeffrey.gordon@nyu.edu)",
        "Connection": "close"
    }
    time.sleep(1) # we need to make sure we wait at list one second between requests

    for i in range(retries):        
        try:
            response = requests.get(
                url=url,
                headers=headers,
                params=params,
                timeout=10  # prevent long hangs
            )
            response.raise_for_status()
            return response.json()

        except requests.exceptions.RequestException:
            if i == retries - 1:
                raise
            time.sleep(2 ** i)  # exponential backoff

def get_artist_id(artist_name):
    artist_search_json =request_musicbrainz(
        "artist",
        {
            "query": f"artist:{artist_name}",
            "fmt": "json",
            "limit": 1
        }
    )

    if artist_search_json['artists']:
        return artist_search_json['artists'][0]['id']
    else:
        return None

def get_artist_details_json(artist_id):
    artist_details_json = request_musicbrainz(
        f"artist/{artist_id}",
        {
            "fmt": "json",
        }
    )

    return artist_details_json

def get_artist_info(artist_name):
    artist_id = get_artist_id(artist_name)

    if not artist_id:
        return None

    artist_details_json = get_artist_details_json(artist_id)

    if not artist_details_json:
        return None

    return ArtistInfo(
        artist_name=artist_name,
        gender=artist_details_json.get('gender'),
        country=artist_details_json.get('country')
    )
```

```python
caffeinate_proc = subprocess.Popen(["caffeinate", "-dims"])

# Check if artists.csv already exists to support resuming
csv_file = 'data/artists.csv'
processed_artists = set()
if os.path.exists(csv_file):
    # Load existing data to skip already processed artists
    df_existing = pd.read_csv(csv_file)
    processed_artists = set(df_existing['artist_name'].values)
    print(f"Found existing file with {len(processed_artists)} artists. Resuming...")
else:
    # Initialize CSV file with headers
    with open(csv_file, 'w', newline='') as f:
        writer = csv.writer(f)
        writer.writerow(['artist_name', 'gender', 'country'])

try:
    for artist_name in relevant_artist_names:

        # Skip if already processed
        if artist_name in processed_artists:
            continue
            
        artist_info = get_artist_info(artist_name)
        gender = None
        country = None

        if artist_info is not None:
            gender = artist_info.gender if artist_info.gender is not None else None
            country = artist_info.country if artist_info.country is not None else None
                
        with open(csv_file, 'a', newline='') as f:
            writer = csv.writer(f)
            writer.writerow([artist_name, gender, country])

        clear_output(wait=True)
        print("Processed artist:", artist_name, "with gender:", gender, "and country:", country)
        print("Total processed artists: ", len(processed_artists))
        
        processed_artists.add(artist_name)

except Exception as e:
    print(f"Error: {e}")
finally:
    caffeinate_proc.terminate()
```
