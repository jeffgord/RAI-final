```python
import numpy as np
import pandas as pd
from dataclasses import dataclass
import requests
import time
import subprocess
```

```python
df_train = pd.read_csv('data/train.csv')
df_train
```

```python
song_ids = df_train['song_id'].unique()
print("number of unique songs in train set: ", len(song_ids))
```

```python
df_songs = pd.read_csv('data/songs.csv')
df_songs
```

```python
relevant_songs = df_songs[df_songs['song_id'].isin(song_ids)]
relevant_songs # interestingly, there are a few songs from the training data that are not included in the songs data
```

```python
relevant_artists = relevant_songs['artist_name'].unique()
print("number of unique artists in relevant songs: ", len(relevant_artists))
```

```python
df_artists = pd.DataFrame(relevant_artists, columns=['artist_name'])
df_artists['gender'] = None
```

```python
@dataclass
class ArtistInfo:
    artist_name: str
    gender: str
    country: str

def request_musicbrainz(endpoint, params):
    headers = {
        "User-Agent": "SchoolProject/1.0 (jeffrey.gordon@nyu.edu)"
    }
    response = requests.get(f"https://musicbrainz.org/ws/2/{endpoint}", headers=headers, params=params)
    return response.json()

def get_artist_id(artist_name):
    artist_search_json =request_musicbrainz(
        "artist",
        {
            "query": f"artist:{artist_name}",
            "fmt": "json",
            "limit": 1
        }
    )

    # avoid rate limiting
    time.sleep(1)

    if artist_search_json['artists']:
        return artist_search_json['artists'][0]['id']
    else:
        return None

def get_artist_details_json(artist_id):
    artist_details_json = request_musicbrainz(
        f"artist/{artist_id}",
        {
            "fmt": "json",
        }
    )

    # avoid rate limiting
    time.sleep(1)

    return artist_details_json

def get_artist_info(artist_name):
    artist_id = get_artist_id(artist_name)

    if not artist_id:
        return None

    artist_details_json = get_artist_details_json(artist_id)

    if not artist_details_json:
        return None

    return ArtistInfo(
        artist_name=artist_name,
        gender=artist_details_json.get('gender'),
        country=artist_details_json.get('country')
    )
```

```python
caffeinate_proc = subprocess.Popen(["caffeinate", "-dims"])

try:
    for idx, row in df_artists.iterrows():
        artist_name = row['artist_name']
        artist_info = get_artist_info(artist_name)
        
        if artist_info is not None:
            df_artists.at[idx, 'gender'] = artist_info.gender
            df_artists.at[idx, 'country'] = artist_info.country

    df_artists.to_csv('data/artists.csv', index=False)
except Exception as e:
    print(f"Error: {e}")
finally:
    caffeinate_proc.terminate()
```

```python
# Save the df_artists DataFrame to CSV
print(f"Saved {len(df_artists)} artists to data/artists.csv")
print(f"Artists with gender info: {df_artists['gender'].notna().sum()}")
print(f"Artists without gender info: {df_artists['gender'].isna().sum()}")
```
